Introducción:
La Alianza del Pacifico es un bloque de cooperación economica compuesto por México, Colombia, Perú y Chile creado en el año 2011 con el fin de fortalecer los lazos comerciales así como promover el crecimiento, el desarrollo y la competitividad de las economías de los países miembro, en el presente proyecto se analizará la evolución de dos indicadores claves de la economía de estos países como lo son el crecimiento del PIB y la tasa de inflación entre 1971 y 2022 usando datos del Banco Mundial con el objetivo de evaluar si a 13 años de la creación de este bloque es posible evidenciar una mayor correlación entre los indicadores económicos seleccionados para el análisis en comparación con años previos y evaluar que tanto ha beneficiado a estos países el pertenecer a este bloque económico en relación al crecimiento del PIB y la evolución de la inflación.

```python
# Se importan las librerias necesarias.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# URL del archivo CSV
url = 'https://raw.githubusercontent.com/Oscjaram/proyectofinal/main/AP_PIB_IPC.csv'

# Leemos los datos desde la URL con pandas
AP_PIB_IPC = pd.read_csv(url)
AP_PIB_IPC.head()
```

```python
import matplotlib.pyplot as plt
```

## Gráfico del Crecimiento del PIB de Colombia entre 1951 y 2022

```python
# Código para Gráfico para el PIB de Colombia
AP_PIB_IPC['Año'] = AP_PIB_IPC['Año'].astype(int)
plt.plot(AP_PIB_IPC['Año'], AP_PIB_IPC['Crecimiento PIB COL'], color='#F1C40F')
plt.title('Crecimiento del PIB de Colombia entre 1951 - 2022')
plt.xlabel('Año')
plt.ylabel('Crecimiento del PIB')
ticks = range(AP_PIB_IPC['Año'].min(), AP_PIB_IPC['Año'].max() + 1, 5)
plt.xticks(ticks)
plt.show()
```

**Análisis sobre la evolución del crecimiento del PIB de Colombia 1971 - 2022:**

Para el caso Colombiano se destaca un crecimiento relativamente estable con periodos de crisis muy puntuales como lo son la crisis financiera presentada en 1999 y la crisis económica generada por la pandemia en el año 2020, en los demás años si bien no se evidencian grandes crisis economicas, tampoco se destacan grandes periodos de crecimiento económico sostenido con lo que si bien la economía no decrece, tampoco crece a gran velocidad, son muy excepcionales los periodos donde se sobrepasa el 7% de crecimiento económico y a esos periodos les suele seguir un periodo de bajo crecimiento.

En relación con la Alianza del pacífico no hay ninguna diferenciación significativa en la evolución del PIB luego de la creación de la misma.


### Conclusiones:
- Solo dos grandes crisis económicas sobresalen en el periodo seleccionado: Crisis financiera en 1999 y crisis por la pandemia en el año 2000
- No se destacan periodos de crecimiento economico acelerado y sostenido, sino más bien un crecimiento con muy pocos picos del 7% seguidos de periodos de bajo crecimiento cercanos al 1% y 2%.
- No hay diferencia alguna en la evolución del crecimiento del PIB luego de la creación de la alianza del pacífico.

## Gráfico de la evolución de la inflación de Colombia entre 1951 y 2022

```python
# Código para gráfico de la evolución de la inflación de Colombia entre 1951 y 2022
AP_PIB_IPC['Año'] = AP_PIB_IPC['Año'].astype(int)
plt.plot(AP_PIB_IPC['Año'], AP_PIB_IPC['IPC COL'], color='#F1C40F')
plt.title('Evolución del IPC de Colombia entre 1951 - 2022')
plt.xlabel('Año')
plt.ylabel('IPC')
ticks = range(AP_PIB_IPC['Año'].min(), AP_PIB_IPC['Año'].max() + 1, 5)
plt.xticks(ticks)
plt.show()
```

**Análisis sobre la evolución la inflacion de Colombia 1971 - 2022:**

Para el caso de la inflación en Colombia se pueden identificar dos grandes periodos, el primero entre 1971 y 1997 en los que las tasas de inflación en el país eran relativamente altas sin llegar a escenarios de hiperinflación vistos en los otros países de la alianza del pacífico, con tasas entre 15% y 35%, el segundo periodo se da entre el año 1998 y 2022 en los que la inflación generalmente se encuentra debajo del 10%, con contados picos como los de los años 2016 y 2022, en relación con la alianza del pacífico no se evidencia ninguan diferenciación dado que la inflación había caido mucho antes de que se diera la cración del bloque.

## Conclusiones:
- Dos grandes periodos identificados, alta inflación entre 1971 y 1997 y una inflación relativamente controlada entre 1998 y 2022.
- No hay diferencia alguna en la evolución de la inflación luego de la creación de la alianza del pacífico.

## Gráfico del Crecimiento del PIB de México entre 1951 y 2022

```python
AP_PIB_IPC['Año'] = AP_PIB_IPC['Año'].astype(int)
plt.plot(AP_PIB_IPC['Año'], AP_PIB_IPC['Crecimiento PIB MEX'],color='#229954')
plt.title('Crecimiento del PIB de Mexico entre 1951 - 2022')
plt.xlabel('Año')
plt.ylabel('Crecimiento del PIB')
ticks = range(AP_PIB_IPC['Año'].min(), AP_PIB_IPC['Año'].max() + 1, 5)
plt.xticks(ticks)
plt.show()
```

**Análisis sobre la evolución del crecimiento del PIB de Mexico 1971 - 2022:**
En el caso Mexicano se observan periodos de crisis ecónomicas profundas seguidos por periodos en los que la economía se recupera a tasas relativamente altas, el periodo de mayor crecimiento se da entre los años 1971 y 1981, luego se da el PIB entra en lo que se conoce como la década perdida etre los 80 y los 90 donde se observan crisis económicas y bajo crecimiento, llegados los 90s se observa un mejor crecimiento económico que sin embargo no esquiva un leve periodo de bajo crecimiento hacia el año 2000, este periodo destaca por tener mayor estabilidad en relación con los demás, las siguientes crisis llegarían con en el año 2009 producto del contagio financiero mundial y la crisis económica desatada por la pandemia en 2020.

En relación con la Alianza del pacífico se observa una mayor estabilidad en el indicador del crecimiento económico del PIB a partir de 2011, situación interrumpida por la crisis de la pandemia.


### Conclusiones:
- En los años seleccionados se observan varias crisis económicas siendo la más profunda la generada por la pandemia.
- El periodo de mayor crecimiento económico se da antes de la decada perdida, luego de esa crisis, Mexico no volvió a crecer a tasas similares.
- La evolución del PIB es un poco más estable luego de la creación de la Alianza del pacifico, a excepción del año de la pandemia.

## Gráfico de la evolución de la inflación de Mexico entre 1951 y 2022

```python
AP_PIB_IPC['Año'] = AP_PIB_IPC['Año'].astype(int)
plt.plot(AP_PIB_IPC['Año'], AP_PIB_IPC['IPC MEX'],color='#229954')
plt.title('Evolución del IPC de Mexico entre 1951 - 2022')
plt.xlabel('Año')
plt.ylabel('IPC')
ticks = range(AP_PIB_IPC['Año'].min(), AP_PIB_IPC['Año'].max() + 1, 5)
plt.xticks(ticks)
plt.show()
```

**Análisis sobre la evolución la inflacion de México 1971 - 2022:**

Para el caso de la inflación en México se pueden identificar dos grandes periodos, el primero entre 1971 y 1997 en los que las tasas de inflación en el país eran altas enfrentando escenarios de hiperinflación, con tasas que superaron el 100% en más de una ocasión, el segundo periodo se da entre el año 1998 y 2022 en los que la inflación generalmente se encuentra debajo del 20%, con contados picos como el del año 2022, cabe destacar que hacia 1993 Mexico elimina 3 ceros a su moneda con lo que la inflación solo se logra controlar a partir de este periodo, en relación con la alianza del pacífico no se evidencia ninguan diferenciación dado que la inflación había caido mucho antes de que se diera la cración del bloque.

## Conclusiones:
- Dos grandes periodos identificados, hiperinflación entre 1971 y 1997 y una inflación relativamente controlada entre 1998 y 2022.
- No hay diferencia alguna en la evolución de la inflación luego de la creación de la Alianza del pacífico.

## Gráfico del Crecimiento del PIB de Perú entre 1951 y 2022

```python
AP_PIB_IPC['Año'] = AP_PIB_IPC['Año'].astype(int)
plt.plot(AP_PIB_IPC['Año'], AP_PIB_IPC['Crecimiento PIB PER'], color='#CB4335')
plt.title('Crecimiento del PIB de Perú entre 1951 - 2022')
plt.xlabel('Año')
plt.ylabel('Crecimiento del PIB')
ticks = range(AP_PIB_IPC['Año'].min(), AP_PIB_IPC['Año'].max() + 1, 5)
plt.xticks(ticks)
plt.show()
```

**Análisis sobre la evolución del crecimiento del PIB de Perú 1971 - 2022:**

Para Perú se pueden observar dos periodos diferenciados en relación a la evolucoión del PIB, el primero se da entre 1971 y 1991 en los que las crisis económicas son relativamente constantes y se acentuan durante la década perdida entre 1980 y 1990, a partir de los años 90 el crecimiento económico de Perú se estabiliza y aunque siguen habiendo periodos de bajo crecimiento la economía crecio a buen ritmo a excepción del año de la pandemia.

En relación con la Alianza del pacífico se observa un menor crecimiento a partir de la creación del bloque económico.


### Conclusiones:
- Dos periodos marcadamente definidos en la evoluición del PIB peruano, entre 1971 y 1991 crisis economicas recurrentes y entre 1991 y 2022 un crecimiento destacable a excepción del año 2020.
- Crecimiento económico más moderado a partir de la creación del bloque.


## Gráfico de la evolución de la inflación de Perú entre 1951 y 2022

```python
AP_PIB_IPC['Año'] = AP_PIB_IPC['Año'].astype(int)
plt.plot(AP_PIB_IPC['Año'], AP_PIB_IPC['IPC PER'],color='#CB4335')
plt.title('evolución del IPC de Perú entre 1951 - 2022')
plt.xlabel('Año')
plt.ylabel('IPC')
ticks = range(AP_PIB_IPC['Año'].min(), AP_PIB_IPC['Año'].max() + 1, 5)
plt.xticks(ticks)
plt.show()
```

**Análisis sobre la evolución la inflacion de Perú 1971 - 2022:**

En Perú se evidencia el mayor caso de hiperinflación de los paises del bloque, llegando a tasas de 7000% en su punto más alto hacia el año 1988, la inflación ha estado bastante controlada durante el resto del tiempo. Debido a la distorción generada por el dato de hiperinflación, al tomar los datos desde 1993 se puede observar mejor el exito peruano al controlar la inflación con tasas menores al 10% ante la ola inflacionaria que recorrio el mundo en 2022.

## Conclusiones:
- Perú tuvo una hiperinflación grave llegando a tasas de 7000%.
- No hay diferencia alguna en la evolución de la inflación luego de la creación de la Alianza del pacífico.

### Conclusiones:
- Diferencias marcadas en el ciclo económico de los países hasta los años 90s.
