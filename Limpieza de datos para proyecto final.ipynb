### Carga de datos
#### Se procede a cargar los datos del PIB y del IPC extraidos del Banco Mundial, estos datos están contenidos en archivos CSV por lo que se usa la librería de Pandas para leerlos.

```python
import pandas as pd
DatosPib = pd.read_csv("data/crec_PIB.csv", engine='python' ,quotechar='"' )
DatosPib
```

```python
DatosIPC = pd.read_csv("data/IPC-BM.csv", engine='python' ,quotechar='"')
DatosIPC
```

### Proceso de limpieza de datos
#### Se evidencia que los datos del PIB están entre comillas, mientras que los datos del IPC no las tienen,  por lo que se procede a crear una función para eliminarlas del dataset del PIB con eso se estandarizan los datos.

```python
#Codigo para eliminar las comillas de los datos del PIB
def remove_quotes(x):
    if isinstance(x, str):
        return x.strip('"')
    else:
        return x
```

```python
#Le aplico la función creada anteriormente para remover las comillas
DatosPib = DatosPib.map(remove_quotes)
```

#### Convierto los datos vacios de las columnas de los años a NAN para poder transformarlos a Float, ya que el PIB y el IPC se expresan en porcentaje.

```python
#Convierto los datos vacios de las columnas de los años a NAN para luego poder transformarlos a float, en el dataset del PIB
cols_to_convert = ['"1960"', '"1961"', '"1962"', '"1963"', '"1964"',
       '"1965"', '"1966"', '"1967"', '"1968"', '"1969"', '"1970"', '"1971"',
       '"1972"', '"1973"', '"1974"', '"1975"', '"1976"', '"1977"', '"1978"',
       '"1979"', '"1980"', '"1981"', '"1982"', '"1983"', '"1984"', '"1985"',
       '"1986"', '"1987"', '"1988"', '"1989"', '"1990"', '"1991"', '"1992"',
       '"1993"', '"1994"', '"1995"', '"1996"', '"1997"', '"1998"', '"1999"',
       '"2000"', '"2001"', '"2002"', '"2003"', '"2004"', '"2005"', '"2006"',
       '"2007"', '"2008"', '"2009"', '"2010"', '"2011"', '"2012"', '"2013"',
       '"2014"', '"2015"', '"2016"', '"2017"', '"2018"', '"2019"', '"2020"',
       '"2021"', '"2022"', '"2023"']
DatosPib[cols_to_convert] = DatosPib[cols_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
#Convierto las columnas del dataset del PIB a float y verifico que el cambio se realizó.
DatosPib[cols_to_convert] = DatosPib[cols_to_convert].astype(float)
DatosPib.dtypes
```

```python
#Convierto los datos vacios de las columnas de los años a NAN para luego poder transformarlos a float, en el dataset del IPC
cols_to_convertipc = ['1960', '1961', '1962', '1963', '1964',
       '1965', '1966', '1967', '1968', '1969', '1970', '1971',
       '1972', '1973', '1974', '1975', '1976', '1977', '1978',
       '1979', '1980', '1981', '1982', '1983', '1984', '1985',
       '1986', '1987', '1988', '1989', '1990', '1991', '1992',
       '1993', '1994', '1995', '1996', '1997', '1998', '1999',
       '2000', '2001', '2002', '2003', '2004', '2005', '2006',
       '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022', '2023']
DatosIPC[cols_to_convert] = DatosIPC[cols_to_convertipc].apply(pd.to_numeric, errors='coerce')
```

```python
DatosIPC[cols_to_convertipc] = DatosIPC[cols_to_convertipc].astype(float)
DatosPib.dtypes
```

### Agrupación de datos
